# Primer paso, extraer los datos de un archivo CSV.

```python
import pandas as pd
extracted_data = pd.read_csv("conjunto_de_datos_iter_00CSV20.csv")
```

# Verificar el contenido.

```python
extracted_data.head()
```

```python
# Borrar renglones donde LOC sea igual a 9999 , 9998 y 0
extracted_data = extracted_data.drop(extracted_data[extracted_data["LOC"]==9999].index)
extracted_data = extracted_data.drop(extracted_data[extracted_data["LOC"]==9998].index)
extracted_data = extracted_data.drop(extracted_data[extracted_data["LOC"]==0].index)
```

# Transformar datos

```python
def tranform(data_frame, value):
    return data_frame.loc[data_frame["ENTIDAD"] == 25, ["ENTIDAD","NOM_ENT","LOC","NOM_LOC","POBTOT","POBFEM","POBMAS","VPH_SINCINT"]]
transformed_data = tranform(data_frame=extracted_data,value="Sinaloa" )
```

# Cargar datos.

```python
def load(data_frame,filename):
    return data_frame.to_csv(filename)
load(data_frame=transformed_data,filename="cleaned_data.csv")
```
